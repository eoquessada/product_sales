```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sales_dataframe = pd.read_excel('Sales_v1.xlsx')
sales_dataframe
```

```python
sales_dataframe['YearMonth'] = sales_dataframe['Year'].astype(str) + '-' + sales_dataframe['Month Number'].astype(str)
sales_dataframe
```

```python
profit_by_yearmonth = sales_dataframe.groupby('Year')['Profit'].sum()
profit_by_yearmonth.plot(kind='line', title='Profit By Year')
```

```python
profit_by_segment = sales_dataframe.groupby('Segment')['Profit'].sum()
profit_by_segment
```

```python
profit_by_segment.plot(kind='bar', title='Profit by Segment')
```

```python
profit_by_product = sales_dataframe.groupby('Product')['Profit'].sum()
profit_by_product.plot(kind='bar', title='Profit by Product')
```

```python
cogs_by_segment = sales_dataframe.groupby('Segment')['Profit'].sum()
cogs_by_segment.plot(kind='bar', title='COGS by Segment')
```
